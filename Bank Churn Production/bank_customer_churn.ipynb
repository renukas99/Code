```python
col_to_scale=['EstimatedSalary','Balance','CreditScore','Tenure','NumOfProducts']

#we have to take minmax scaler from sklear it will convert value to 0 and 1

from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
```

```python
df2[col_to_scale]=scaler.fit_transform(df2[col_to_scale])
df2=df2.drop('Age',axis='columns')
df2
```

```python
x=df2.drop('Exited',axis='columns')
y=df2.Exited
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test=train_test_split(x, y, test_size=0.2,random_state=5)
```

```python
x_train.shape
```

```python
import tensorflow as tf
from tensorflow import keras


model = keras.Sequential([
    keras.layers.Dense(11, input_shape=(11,), activation='relu'),
    keras.layers.Dense(1, activation='sigmoid')
])

#Loss is binary creoss entropy bcz we have 0 & 1 value of output
model.compile(optimizer='adam',
              loss='binary_crossentropy',
              metrics=['accuracy'])

model.fit(x_train, y_train, epochs=100)
```

```python
model.evaluate(x_test, y_test)
```

```python
yp=model.predict(x_test)
yp[0:5]
```

```python
y_pred =[]

for element in yp:
  if element >0.5:
    y_pred.append(1)
  else:
    y_pred.append(0)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred))
```

```python
import seaborn as sn
cm= tf.math.confusion_matrix(labels=y_test, predictions=y_pred)

plt.figure(figsize = (10,7))
sn.heatmap(cm, annot=True, fmt='d')
plt.xlabel('Predicted')
plt.ylabel('Truth')
```
